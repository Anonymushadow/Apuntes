```python
años = ["2011", "2012", "2013", "2014", "2015", "2016", "2017", "2018", "2019", "2020"]

nivel1 = np.random.rand(10) * 100
nivel2 = np.random.rand(10) * 200 + 100
nivel3 = np.random.rand(10) * 300 + 200 
nivel4 = np.random.rand(10) * 400 + 300
nivel5 = np.random.rand(10) * 500 + 400
```

```python
#introducir series de datos con color, etiqueta y estilos
#plot(eje x, eje y, nombre, coor, marcadores, tipo de linea)
plt.plot(años, nivel1, label = "Nivel 1", color = "#00cc00", marker = "<", linestyle = "-")
plt.plot(años, nivel2, label = "Nivel 2", color = "#cc9900", marker = "*", linestyle = "--")
plt.plot(años, nivel3, label = "Nivel 3", color = "#009999", marker = "x", linestyle = ":")
plt.plot(años, nivel4, label = "Nivel 4", color = "#ff0000", marker = "+", linestyle = "-.")
plt.plot(años, nivel5, label = "Nivel 5", color = "#0000ff", marker = ".", linestyle = " ")

#Activar leyendas
plt.legend()         
         
#Titulo del grafico
plt.title("Examen de certificacion")

#Etiquetas de lo ejes
plt.xlabel("Años")
plt.ylabel("Puntaje")

#Personalizar eje de puntaje de 0 a 1000 en incremento de 50
plt.yticks(np.arange(0, 1001, 50))

#Activar cuadricula
plt.grid()

#Activar marcas menores
plt.minorticks_on()

plt.show()
```

```python
eje_x = np.arange(10)

plt.bar(eje_x, nivel5, width = 1/5)
plt.bar(eje_x + 0.2, nivel4, width = 1/5)
plt.bar(eje_x + 0.4, nivel3, width = 1/5)
plt.bar(eje_x + 0.6, nivel2, width = 1/5)
plt.bar(eje_x + 0.8, nivel1, width = 1/5)

plt.show()
```

```python
matriz_confusion = np.array([[22, 6, 3], [6, 20, 4], [5, 1, 24]])

#señal 8 segundos con frecuencia de muestra de 128Hz
#suma de 3 señales seno

fs = 128
time = np.arange(0, 8, 1/fs)
x1 = np.sin(2 * np.pi * 2 * time) #seno 2Hz
x2 = np.sin(2 * np.pi * 5 * time) #seno 5Hz
x3 = np.sin(2 * np.pi * 8 * time) #seno 8Hz
sig = x1 + x2 + x3

plt.plot(time, sig)
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.grid()
plt.show()
```

```python
plt.pcolormesh(matriz_confusion, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
#colores mas oscuros = valores mas bajos
plt.imshow(matriz_confusion)
plt.colorbar()
plt.show()
```

```python
#3d
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure() #definir la figura
ax = Axes3D(fig) #donde se dibuja la figura

x = np.linspace(-4, 4, 50)#variable independiente, desde -4 hasta 4 con 50 puntos
y = np.linspace(-4, 4, 50)

#funcion que recibe x e Y y devuelve el seno de el cuadrado de X + el cuadrado de Y
def z(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

ax.plot(x, y, z(x, y)) #dibuja X, Y y Z
plt.show()
```

```python
fig = plt.figure() #definir la figura
ax = Axes3D(fig) #donde se dibuja la figura

x = np.linspace(-4, 4, 50)
y = np.linspace(-4, 4, 50)
x, y = np.meshgrid(x, y) #pasamos coordenadas y devulve matrices de coordenadas

#funcion que recibe x e Y y devuelve el seno de el cuadrado de X + el cuadrado de Y
def z(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

ax.plot_surface(x, y, z(x, y)) #dibuja X, Y y Z
plt.show()
```
