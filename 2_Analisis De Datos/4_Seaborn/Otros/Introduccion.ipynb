```python
#crea un dataframe y pasa el tiempo y los valores con sus acumulacion sumada a un array
from numpy.random import randn
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()), index = None)
df
```

```python
_ = sns.relplot(x = "time", y = "value", kind = "line", data = df, sort = True)
```

```python
df = pd.DataFrame(randn(500, 2).cumsum(axis = 0), columns = ["time", "value"]) 
df
```

```python
#ordenado
_ = sns.relplot(x = "time", y = "value", kind = "line", data = df, sort = True)
```

```python
#desordenado
_ = sns.relplot(x = "time", y = "value", kind = "line", data = df, sort = False)
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
#el espectro que se ve es la probabilidad, el resultado no es 100% preciso, puede estar en cualquier parte del espectro
_ = sns.relplot(x = "timepoint", y = "signal", kind = "line", data = fmri)
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", kind = "line", data = fmri, ci = False)
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", kind = "line", data = fmri, ci = "sd")
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", estimator = None, kind = "line", data = fmri)
```

```python
#hue determina qué columna del marco de datos debe usarse para la codificación de color
_ = sns.relplot(x = "timepoint", y = "signal", hue = "event", kind = "line", data = fmri)
```

```python
#style es una Variable de agrupación que producirá elementos con diferentes estilos
#style usa linea punteada y hue no
_ = sns.relplot(x = "timepoint", y = "signal", hue = "region", kind = "line", data = fmri, style = "event")
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", hue = "region", kind = "line", data = fmri, style = "event", markers = True)
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", hue = "region", kind = "line", data = fmri, style = "event", markers = True, dashes = False)
```

```python
_ = sns.relplot(x = "timepoint", y = "signal", units = "subject", kind = "line", estimator = None, hue = "region", data = fmri)
```

```python
#toma solo los valores de fmri que su evento sea igual a strim
_ = sns.relplot(x = "timepoint", y = "signal", units = "subject", kind = "line", estimator = None, hue = "region", data = fmri.query("event == 'stim'"))
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
dots
```

```python
_ = sns.relplot(x = "time", y = "firing_rate", data = dots, kind = "line", hue = "coherence")
```

```python
_ = sns.relplot(x = "time", y = "firing_rate", data = dots, kind = "line", hue = "coherence", style = "choice")
```

```python
palette = sns.cubehelix_palette(light = 0.8, n_colors = 6)
_ = sns.relplot(x = "time", y = "firing_rate", data = dots, kind = "line", hue = "coherence", style = "choice", palette = palette)
```

```python
#cambiar el tamaño de las lineas
_ = sns.relplot(x = "time", y = "firing_rate", size = "coherence", sizes = (1, 5), data = dots, kind = "line", hue = "coherence", style = "choice", palette = palette)
```
