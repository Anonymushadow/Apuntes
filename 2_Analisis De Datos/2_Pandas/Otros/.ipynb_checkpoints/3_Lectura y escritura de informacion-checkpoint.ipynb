```python
sr = pd.Series([1, 2, 3, 4, 5, 6])
sr
```

```python
def es_par(elemento):
    if elemento % 2 == 0:
        return f'Par: {elemento}'
    else:
        return f'Impar: {elemento}'
```

```python
sr.map(es_par)
```

```python
df = pd.DataFrame(np.arange(16).reshape(4, 4))
df
```

```python
df.applymap(es_par)
```

```python
df = pd.DataFrame(np.arange(16).reshape(4, 4))
df
```

```python
def es_suma_par(elemento):
    x = np.sum(elemento)
    if x % 2 == 0:
        return f'Suma Par: {x}'
    else:
        return f'Suma Impar: {x}'
```

```python
df.apply(es_suma_par, axis = 0)
```

```python
df.apply(es_suma_par, axis = 1)
```

```python
peliculas = pd.DataFrame({"Año": [2014, 2014, 2013, 2013],
                         "Valoraciones": [6, None, 8.5, None],
                         "Presupuesto": [160, 250, 100, None],
                         "Director": ["Gretch Edwards", "Peter Jackson", "Martin Scorsese", "Pedro Almodovar"],
                         "Titulo": ["Godzilla", "El hobbit III", "El lobo de Wall Street", "Gravity"]})
peliculas
```

```python
directores = pd.DataFrame({"Director": ["Gretch Edwards", "Martin Scorsese", "Pedro Almodovar"],
                         "Año Nacimiento": [1975, 1942, 1949],
                         "Nacionalidad": ["England", "USA", "Spain"]})
directores
```

```python
pd.merge(peliculas, directores)
```

```python
#de peliculas tomas Director y lo une con Nombres de directores,quedando solo una
directores.columns = ["Nombre", "Nacimiento", "Nacionalidad"]
pd.merge(peliculas, directores, left_on = "Director", right_on = "Nombre")
```

```python
pd.merge(peliculas, directores, left_on = "Director", right_on = "Nombre", how = "right")
```

```python
pd.merge(peliculas, directores, left_on = "Director", right_on = "Nombre", how = "left")
```

```python
pd.merge(peliculas, directores, left_on = "Director", right_on = "Nombre", how = "outer")
```

```python
peliculas = pd.DataFrame({"Año": [2014, 2014, 2013, 2013],
                         "Valoraciones": [6, None, 8.5, None],
                         "Presupuesto": [160, 250, 100, None],
                         "Director": ["Gretch Edwards", "Peter Jackson", "Martin Scorsese", "Pedro Almodovar"],
                         "Titulo": ["Godzilla", "El hobbit III", "El lobo de Wall Street", "Gravity"]})
peliculas

directores = pd.DataFrame({"Director": ["Gretch Edwards", "Martin Scorsese", "Pedro Almodovar"],
                         "Año Nacimiento": [1975, 1942, 1949],
                         "Nacionalidad": ["England", "USA", "Spain"],
                          "Valoracion": [6, 7, 8]})
directores

pd.merge(peliculas, directores, left_on = "Director", right_on = "Director")
```

```python
pd.merge(peliculas, directores, left_on = "Director", right_on = "Director", how = "outer", suffixes = ("peli", "dire"))
```
